```python
#Attention à la relation kcal par personne sur le tableau dispo_alimentaire. Cette serie
#ne nous permet pas de connaitre le nombre reel de la disponibilite calorique dans un
#pays. Pour cela, il faut, d'abord, réaliser un groupby par pays sur le df dispo_alimentaire
#On a ainsi un nouveau df "dispo_alimentaire_2017"
dispo_alimentaire_2017=dispo_alimentaire[['Zone','Produit','Disponibilité alimentaire (Kcal/personne/jour)']]
dispo_alimentaire_2017=dispo_alimentaire_2017.groupby(by='Zone').agg({'Disponibilité alimentaire (Kcal/personne/jour)':sum}).sort_values(by='Disponibilité alimentaire (Kcal/personne/jour)', ascending=False)
dispo_alimentaire_2017
```

```python
#Ensuite on filtre les données sur le df population on choisissant l'année cicle 2017
#On a un nouveau df "dispo_alimentaire_pop_2017"
dispo_alimentaire_pop_2017=(population[(population['Année']=='2017')])
dispo_alimentaire_pop_2017
```

```python
#On fait une jonction des dfs afin d'avoir sur un seul tableau l'info de calories
#produite par personne et la population total de chaque pays
#Un nouveau df apparait: merge_dispo_pop_2017
merge_dispo_pop_2017=dispo_alimentaire_pop_2017.merge(dispo_alimentaire_2017, on = ['Zone'])
merge_dispo_pop_2017
```

```python
#Maintenant on multiplie la population d'un pays avec la disponibilite alimentaire
#calorique par personne.
#La nouvelle serie "Total" affichera la disponibilite journalière total cette fois-ci tentant en
#compte la population de chaque pays en 2017
merge_dispo_pop_2017['Total'] = merge_dispo_pop_2017['Disponibilité alimentaire (Kcal/personne/jour)']*merge_dispo_pop_2017['Valeur']
merge_dispo_pop_2017
```

```python
#On fait la somme de la série totale afin de connaitre le volume reel produit par jour
dispo_kcal_jour=round(merge_dispo_pop_2017['Total'].sum())
dispo_kcal_jour
```

En 2017 le monde a produit environs **21 milliards de calories par jour**

```python
#On trouve le volume annuel
dispo_kcal_an=round(dispo_kcal_jour)*365
dispo_kcal_an
```

Cela represente environs **7 trillions de calories produite en 2017**

```python
#On divisant le volume annuel de calories produite en 2017 on peut, enfin, trouver le
#nombre de personnes qui auraient pu, en théorie, être nourries
#On considere qu'un être humain a besoin de 912500kcal par an (ou 2500kcal par jour) pour vivre, selon l'OMS
nombre_personnes_nourries_kcal=round(dispo_kcal_an/912500)*1000
nombre_personnes_nourries_kcal
```

Le nombre total de calories produites en 2017 aurait pu nourrir **8,3 milliards de personne**

```python
#Maintenant on calcule la proportion de personne dans une échelle mondiale
prop_nombre_personnes_nourries=(nombre_personnes_nourries_kcal*100/pop_2017)
prop_nombre_personnes_nourries
```

Le chiffre au-dessus represente **110% de la population mondiale**
